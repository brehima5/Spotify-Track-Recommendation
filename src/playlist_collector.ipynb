```python
import pandas as pd
import numpy as np
import spotipy
from spotify_client import sp
```

## Function to fetch playlist name and basic information from spotify

```python
def fetch_playlists(playlist_ids):
    playlists = [] # i create a list to store my playlist information

# for each playlist_id inside my list, i will loop through and get its info
    for playlist_id in playlist_ids:
        try:
            playlist = sp.playlist(playlist_id)

            playlists.append({
               "Playlist_id": playlist["id"],
               "Playlist_name": playlist['name']
            })
        except Exception as error:
            print("could not fetch playlist")

    return playlists

# trying a public id
fetch_playlists(['3U3LMqLVnwTGlAjiuwafF9'])
```

## Creating a function to get the tracks inside each playlist, that will serve for track analysis

```python
def get_all_playlist_tracks(playlist_id):
    tracks = []
    results = sp.playlist_tracks(playlist_id)

    while results:
        for item in results["items"]:
            track = item["track"]

            if track:  # sometimes None if removed
                tracks.append({
                    "track_name": track["name"],
                    "artist": track["artists"][0]["name"],
                    "track_id": track["id"]
                })

        results = sp.next(results) if results["next"] else None

    return tracks

```
